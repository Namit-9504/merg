## DBSCAN Clustering Anomaly Detection

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
X,y=make_circles(n_samples=750,noise=0.1,factor=0.3)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
DBS=DBSCAN(eps=0.10)
```

```python
DBS.fit_predict(X)
```

```python
DBS.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=DBS.labels_)
```

## Local Outlier Factor Anamoly Detection

```python
from sklearn.datasets import make_circles
X,y=make_circles(n_samples=250,noise=0.1,factor=0.3)
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof=LocalOutlierFactor(n_neighbors=5)
```

```python
lof.fit_predict(X)
```

```python
outlier_scores= - lof.fit_predict(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=outlier_scores)
```
