## EDA With Red Wine Data



Data Set Information:

The two datasets are related to red and white variants of the Portuguese "Vinho Verde" wine.  Due to privacy and logistic issues, only physicochemical (inputs) and sensory (the output) variables are available (e.g. there is no data about grape types, wine brand, wine selling price, etc.).

These datasets can be viewed as classification or regression tasks. The classes are ordered and not balanced (e.g. there are many more normal wines than excellent or poor ones). Outlier detection algorithms could be used to detect the few excellent or poor wines. Also, we are not sure if all input variables are relevant. So it could be interesting to test feature selection methods.


Attribute Information:

Input variables (based on physicochemical tests):
- 1 - fixed acidity
- 2 - volatile acidity
- 3 - citric acid
- 4 - residual sugar
- 5 - chlorides
- 6 - free sulfur dioxide
- 7 - total sulfur dioxide
- 8 - density
- 9 - pH
- 10 - sulphates
- 11 - alcohol

Output variable (based on sensory data):
- 12 - quality (score between 0 and 10)

```python
# Explore More Visulazition
fig,axis=plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='avgrage',bins=30,color='g',kde=True)
plt.subplot(122)
sns.histplot(data=df,hue='gender',x='avgrage',kde=True,bins=30)
plt.show()
```

##### Insights or Observation
- female students makes good marks as compairs to male students

```python
# Explore More Visulazition
fig,axis=plt.subplots(1,3,figsize=(25,6))
plt.subplot(141)
sns.histplot(data=df,x='avgrage',bins=30,hue='lunch',kde=True)
plt.subplot(142)
sns.histplot(data=df[df.gender=='female'],hue='lunch',x='avgrage',kde=True)
plt.subplot(143)
sns.histplot(data=df[df.gender=='male'],hue='lunch',x='avgrage',kde=True)
plt.show()
```

##### Insights or Observation
- Standard Lunch helps perform well in exam in case of male and female

```python
fig,axis=plt.subplots(1,3,figsize=(25,6))
plt.subplot(141)
sns.histplot(data=df,x='avgrage',bins=30,hue='parental_level_of_education',kde=True)
plt.subplot(142)
sns.histplot(data=df[df.gender=='female'],hue='parental_level_of_education',x='avgrage',kde=True)
plt.subplot(143)
sns.histplot(data=df[df.gender=='male'],hue='parental_level_of_education',x='avgrage',kde=True)
plt.show()
```

#####  Insights
- In general parent's education don't help student perform well in exam.
- 2nd plot shows that parent's whose education is of associate's degree or master's degree their male child tend to perform well in exam
- 3rd plot we can see there is no effect of parent's education on female students.

```python
fig,axis=plt.subplots(1,3,figsize=(25,6))
plt.subplot(141)
sns.histplot(data=df,x='avgrage',bins=30,hue='race_ethnicity',kde=True)
plt.subplot(142)
sns.histplot(data=df[df.gender=='female'],hue='race_ethnicity',x='avgrage',kde=True)
plt.subplot(143)
sns.histplot(data=df[df.gender=='male'],hue='race_ethnicity',x='avgrage',kde=True)
plt.show()
```

#####  Insights
- Students of group A and group B tends to perform poorly in exam.
