```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()  ## this helps us to render the graph inside the notebook
```

```python
from bokeh.plotting import figure, output_file , show 
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
output_file("test.html")
p = figure(title = 'test flower')
p.xaxis.axis_label = 'petal length'
p.xaxis.axis_label = 'petal width'
p.circle(flowers['petal_length'],flowers['petal_width']) # beacuse of 'circle it is plotting circles on graph 
show(p)
```

```python
output_file("test.html")
p = figure(title = 'test flower')
p.xaxis.axis_label = 'petal length'
p.xaxis.axis_label = 'petal width'
p.line(flowers['petal_length'],flowers['petal_width']) # if we want to change in graph pattern the we need to change here
show(p)
```

```python
x=[1,2,3,4]
y=[6,5,4,1]
output_file("line.html")
p = figure(title = "line graph")
p.line(x,y)
show(p)
```

```python
x=[1,2,3,4]
y=[6,5,4,1]
output_file("line.html")
p = figure(title = "line graph")
#if we want to increase the size to dots then we will pass size
# if we want to change a color then we use 'fill_color
# if we want to pass some data inside graph then we will use 'legend_label'
p.scatter(x,y,size=20,fill_color = 'red',legend_label='this is my data point') 
show(p)
```
