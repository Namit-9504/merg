##  Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)  ## it takes 1,1 value in series and plot it 
```

```python
## here "c=" means colour of dots and r is a html code for red color we can find any color code on google
## alpha is used to control a color intensity or we say color darkness max of alpha is 1
# marker is used to change a dot into our given chareter
plt.figure(figsize=(6,2)) ## it used to adjust fig size
plt.scatter(x,y , c="#00ff00",alpha=0.8,marker="x") 
plt.xlabel("This is my X axis")  
plt.ylabel("This is my Y axis ")  
plt.title("This is a title of graph") 
plt.grid() ## it gives the square grid at backside of graph
```

```python
a = np.linspace(1,10,100)
b = np.sin(a)
```

```python
a
```

```python
b
```

```python
plt.scatter(a,b)
```

```python
## plot fun can plot a graph where scatter fun plot only points
# "--r" gives a dash dash line in red colour (r is colur code)
plt.figure(figsize=(8,2))
plt.plot(a,b,'--r')
plt.xlabel("this is my x data ")
plt.ylabel("this is my y data ")
plt.title("This is my sin graph")
plt.grid()
```

```python
m=['x','y','a','y','z']
n=np.random.rand(5)
```

```python
m
```

```python
n
```

```python
#"plt.bar" is used to bar graph.Inside plt.bar to give colour we need to use "color='colour_code'"
plt.figure(figsize=(6,2))
plt.bar(m,n,color="#09ad41")
plt.xlabel("representing my cat data ")
plt.ylabel("representing my num data")
plt.title("this is my bar graph")
plt.grid()
```

```python
plt.figure(figsize=(6,2))
plt.barh(m,n,color="#09ad41") ## 'barh' is used to plot graph horizontally i.e on y axis = data on x axis and x axis = data on y axis
plt.xlabel("representing my cat data ")
plt.ylabel("representing my num data")
plt.title("this is my bar graph")
plt.grid()
```

```python
data = [1,2,3,3,4,4,5,5,6]
```

```python
plt.figure(figsize=(6,3))
plt.hist(data,color="#09ad41") ## to plot frequency on y axis and data on x axis # hist=histrogram
plt.xlabel("representing unqine value")
plt.ylabel("representing frequency")
plt.title("Histrogram Example")
plt.grid()
plt.savefig("test.png") ### we san save fig in any type of file i.e. pdf, jpg,png...etc
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig= plt.figure() ## for 3d graph we need to create a figure 
ax = fig.add_subplot(projection = '3d') ## then we need to assign our 3d opretion
ax.scatter(x,y,z,c="r",alpha=.8)
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
## zlable is not avaialable
plt.show()
```
