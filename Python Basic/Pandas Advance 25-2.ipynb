```python
data1 = pd.DataFrame({'key1':(1,2,3,4,5),
                      'key2':(4,5,6,7,8),
                      'key3':(6,7,8,9,7)
})
```

```python
data2 = pd.DataFrame({'key1':(1,26,63,4,5),
                      'key4':(14,5,60,70,8),
                      'key5':(26,76,6,7,56)
})
```

```python
data1
```

```python
data2
```

```python
## mearge operation checks the common column and checks its common values by row wise 
# if value is common then it will put all the row
# if in common column there is a different row then it will delete this row 
```

```python
pd.merge(data1,data2)  ## in merge  by defult how=inner
```

```python
##  it will always assume 1st data set = left and 2nd data set = right 
## in how=left it will keep 1st data set as it is that is left data set as it is 
## and it will check the values in 2nd data set for common values in a common column
```

```python
pd.merge(data1,data2,how='left',on='key1')
```

```python
pd.merge(data1,data2,how='left') 
```

```python
## as we know the fun of how=left 
## if we want to check on the basis of a specific column in data set 1 and data set2 
## i.e. as we check on the basis of common column in a how = left 
## like that here we are checking on the basis of selected columns in data set 1 and data set2 
```

```python
pd.merge(data1,data2,how='left',left_on= 'key2',right_on='key4')
```

```python
##  it will always assume 1st data set = left and 2nd data set = right 
## in how=right it will keep 2nd data set as it is that is right data set as it is 
## and it will check the values in 1st data set for common values in a common colum
```

```python
pd.merge(data1,data2,how='right')
```

```python
##  it will always assume 1st data set = left and 2nd data set = right 
## in case of how = outer
## it will written all the values in a common column 
## and if the value in common column is same then it will written all the values in data set 1 and 2 
## but if the value in common column is not same then it will written NaN 
```

```python
pd.merge(data1,data2,how='outer')
```

```python
##  it will always assume 1st data set = left and 2nd data set = right
## in case of how = outer 
## it will write all the possible cases for cobination
## i.e. it takes 1st row from 1st data set and combine it with all the rows present inside a 2nd data set 
## after that it will take a 2nd row from 1st data set and again combine it with all rows in 2nd data set 
## moveing forward after the last row of 1st data set it will take 1st row of 2nd data set and combines with all rows in 1st data set
## it goes till last row of 2nd data set"
```

```python
pd.merge(data1,data2,how='cross')
```

```python
df4=pd.DataFrame({'a':[1,2,3,4,5,6,7,8,9]})
```

```python
df4
```

```python
### rolling helps us to go to next row  and the window = 1 is means that it takes only one row for caluation
### i.e at a condition of windows = 2 it takes 2 rows for calculation 2 row are the one row and there previous one 
## if previsous row is not available then it writtens NaN 
### for condition windows = n it takes from that row previous nth rows
### whatever operation we will give after windows () that will be performed for that no of rows for whole data frame
```

```python
df4.rolling(window=1).mean() 
```

```python
df4.rolling(window=2).mean()
```

```python
df4.rolling(window=3).mean()
```

```python
df4.rolling(window=3).min()
```
