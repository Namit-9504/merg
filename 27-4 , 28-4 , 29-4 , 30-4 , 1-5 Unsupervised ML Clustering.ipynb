## K Means Practical Implementation

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=23)
```

```python
X
```

```python
y ### in unsupervised meachine learning we are not using target variables
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
## Elbow method for k value
```

```python
wcss=[]
for k in range (1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
### Elbow curve 
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++')
```

```python
y_lables=kmeans.fit_predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_lables)
```

```python
### Knee locator
!pip install kneed
```

```python
from kneed import KneeLocator
```

```python
k1=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
k1.elbow
```

```python
### silhoutte score
from sklearn.metrics import silhouette_score
```

```python
silhouette_cofficients=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train)
    score=silhouette_score(X_train,kmeans.labels_)
    silhouette_cofficients.append(score)
```

```python
silhouette_cofficients
```

```python
plt.plot(range(2,11),silhouette_cofficients)
```

##  Agglomerative Clustering Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
```

```python
X_scaled=scalar.fit_transform(df)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
pca_scaled=pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
## construct dendogram.
## Theroitical Implementation
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(20,7))
plt.title("Dendogram")

## creating dendogram
sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.title("dendrogram")
plt.xlabel("sample index")
plt.ylabel("Eucledain Distance")


### by ploting this graph we get the best values for our K i.e. n_clusters value
### as we see in therory we need to find longest vertical line so that no any horizontal line cut thats line
```

```python
from sklearn.cluster import AgglomerativeClustering
Alc=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
Alc.fit(pca_scaled)
```

```python
Alc.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=Alc.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]

for k in range(2,11):
    aggo=AgglomerativeClustering(n_clusters=k,affinity='euclidean',linkage='ward')
    aggo.fit(X_scaled)
    score=silhouette_score(X_scaled,aggo.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of clusters")
plt.ylabel("Silhotte_coefficient")
plt.show()
```

## DBSCAN Clustering Implementation

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons ### it will create non linear points
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y=make_moons(n_samples=250,noise=0.10)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
dbs=DBSCAN(eps=0.3)
```

```python
dbs.fit(X_scaled)
```

```python
dbs.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbs.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
score=silhouette_score(X,dbs.labels_)
```
